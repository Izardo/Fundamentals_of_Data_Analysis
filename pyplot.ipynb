## Basic plotting functions
***

`plot()`
- plots y against x in the form of a line (as default) or another data marker which can be specified as a parameter. 

`ylabel()` & `xlabel()`
- Sets a label for the axis.

`show()`
- Displays the open figures. 

`plt` is the standard alias for matplotlib.pyplot

```python
# Plot array. 
plt.plot([1, 2, 3, 4])

# Set y label.
plt.ylabel('Y AXIS')

# Show plot.
plt.show()
```

Only one array was passed through the plt.plot() function, pyplot assumes that array to be the y values and based on those it generates the x values automatically. The reason the starting y value is at zero is, of course, because python is a zero based indexing language - meaning its range always starts at zero as default.

<br>

### Passing in both x and y values

```python
# Specifying x and y values.
plt.plot([5, 10, 15, 20], [1, 2, 4, 8])
plt.ylabel('y label')
plt.xlabel('x label')
plt.show()
```

<br>

### Style

As mentioned earlier, the plot() function also takes in a third parameter to specify data marker style. <br>
- The data marker styles are specified as a string containing two values <br>
- The first character denotes the colour (i.e, g = green etc.)
- The second character denotes the shape (i.e, ^ = triangle, - = line etc) 

The following code will generate a random sample using numpy's arange() function to plot various functions. [https://matplotlib.org/stable/tutorials/introductory/pyplot.html#sphx-glr-tutorials-introductory-pyplot-py]

```python
# Generate random range from 0 (inclusive) to 10 (exclusive) in intervals of 0.5
x = np.arange(0, 10, 0.5)

# Plot line with red line, blue square, green triangles, and yellow circles. 
plt.plot(x, x, 'r--', x, x**2, 'bs', x, x**3, 'g^', x, x**2.5, 'yo')
plt.show()
```

#### Default styles


It is good to know that there are around thirty plot styles that can be set as the default.  Have a look at the plot below to see the various plot styles. 

The following code was adapted from [here](https://www.dunderdata.com/blog/view-all-available-matplotlib-styles).

```python
# https://www.dunderdata.com/blog/view-all-available-matplotlib-styles

# Produce a random sample.
x = np.arange(-2, 8, .1)

# Y values
y = .1 * x ** 3 - x ** 2 + 3 * x + 2

# COMMENT LATER.
fig = plt.figure(dpi=100, figsize=(9, 18), tight_layout=True)
default_styles = ['default'] + plt.style.available
for i, style in enumerate(default_styles):
    with plt.style.context(style):
        ax = fig.add_subplot(10, 3, i + 1)
        ax.plot(x, y)
    ax.set_title(style)
```

```python
# View settings for the default plots.
plt.style.library['seaborn-darkgrid']
```

<br>

- To set a default style: `plt.style.use` <br>
- To reset the default to the default style: `plt.style.use('default')`

### Creating subplots with categorical data
***

<br>

The <b>`subplots()`</b> function <b>specifies the plot positions</b> with integers: 

<br>

`subplots(nrows, ncols, index)`

- `nrows` : the row to place plot<br>
The example below has one row.
  
- `ncols` : the column to place plot<br>
The subplot width.

- `index` : the index the plot will be placed<br>
index starts at the upper left corner and increments to the right i.e., upper left index is one, plot to its right is 2 etc.

A three-digit integer without the commas can also be passed through, see the example below. It does the very same as with commas.

```python
#Y values.
y = ['Low', 'Medium', 'High']

# X values.
x = [10, 50, 200]

# Set figure size and number of subplots.
plt.figure(figsize=(12, 3))

# Position the subplot & create bar chart.
plt.subplot(1,3,1)
plt.bar(y, x)

# Using the three integers for specifying subplot position
# & create a scatter plot.
plt.subplot(132)
plt.scatter(y, x)

# Specify subplot position & create a simple plot. 
plt.subplot(133)
plt.plot(y, x)

# Give figure a title. 
plt.suptitle('Categorical Data')
plt.show()
```

### set_antialised()
***

<br>

<i>From [wikipedia](https://en.wikipedia.org/wiki/Spatial_anti-aliasing):</i><br>
&nbsp;&nbsp;&nbsp;&nbsp;"In computer graphics, anti-aliasing improves the appearance of "jagged" polygon edges, or "jaggies", so they are smoothed out on the screen."


<img src='Images/antialiase.gif' style="width:300px">
<i>Close up of antialised v. non-antialiased line</i>

<br>
<br>

The two plots below with `set_antialiased()` takes a Boolean value. By setting the value to True (turn on antialising), it visually improves the line and thus, the overall aesthetic of the plot.

```python
# y-values.
y = [1, 2, 3, 4]
# x-values
x = [5, 10, 30, 90]

# Plot line.
line, = plt.plot(y, x, 'g-')

# Turn on antialiasing.
line.set_antialiased(True)
```

```python
# Plot line.
line, = plt.plot(y, x, 'g-')

# Turn off antialiasing.
line.set_antialiased(False) 
```
